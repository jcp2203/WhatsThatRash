# Creating the training, cross-validation and testing dataset


```python
import numpy as np
import pandas as pd
import os, glob

from sklearn.cross_validation import train_test_split
```

```python
#Give the path of the final images
dir_finim = 'C:\\Users\\Valentin\\Documents\\Columbia University\\Courses\\Data Science for Industry\\Final project\\200with mask'
```

```python
os.listdir(dir_finim)
```

```python
#Reading all the images
i = 0

all_im = []
label = []

for filename in glob.glob(dir_finim + '/*/*.jpg'):
    imageID = filename.split('\\')[-1]
    all_im.append([imageID,filename])
    
    if (filename.split('\\')[-2] == 'ATO'):
        label.append(1)
    if (filename.split('\\')[-2] == 'contact'):
        label.append(2)
    if (filename.split('\\')[-2] == 'MEL'):
        label.append(3)
    if (filename.split('\\')[-2] == 'mim'):
        label.append(4)
    if (filename.split('\\')[-2] == 'PIT'):
        label.append(5)
    if (filename.split('\\')[-2] == 'PSO'):
        label.append(6)
    if (filename.split('\\')[-2] == 'RIN'):
        label.append(7)

all_im_DF = pd.DataFrame(all_im)
all_im_DF.columns = ['ID','Images Path']

all_im_DF.head()
```

```python
#Dividing the data in Train and Test
Xtrain, Xtest, ytrain, ytest = train_test_split(all_im_DF, label, test_size=0.2, random_state=42)
```
